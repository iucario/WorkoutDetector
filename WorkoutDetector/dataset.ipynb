# Dataset Info and Visualization

```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_txt = '../data/Workouts/train.txt'
test_txt = '../data/Workouts/test.txt'
val_text = '../data/Workouts/val.txt'

classes = [
    'front_raise', 'pull_up', 'squat', 'bench_pressing', 'jumping_jack', 'situp',
    'push_up', 'battle_rope', 'exercising_arm', 'lunge', 'mountain_climber']

with open(train_txt, 'r') as f:
    train_list = f.readlines()
    train_dict = dict([(c, 0) for c in classes])
    for line in train_list:
        c = int(line.strip().split(' ')[2])
        train_dict[classes[c]] += 1

with open(test_txt, 'r') as f:
    test_list = f.readlines()
    test_dict = dict([(c, 0) for c in classes])
    for line in test_list:
        c = int(line.strip().split(' ')[2])
        test_dict[classes[c]] += 1

with open(val_text, 'r') as f:
    val_list = f.readlines()
    val_dict = dict([(c, 0) for c in classes])
    for line in val_list:
        c = int(line.strip().split(' ')[2])
        val_dict[classes[c]] += 1
```

```python
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (3, 2)
plt.rcParams['figure.dpi'] = 300
sns.barplot(x=list(range(len(classes))), y=list(train_dict.values()), color='#0066ff')
plt.xlabel('Classes')
plt.ylabel('Number of videos')
plt.title('Train')
plt.show()

sns.barplot(x=list(range(len(classes))), y=list(val_dict.values()), color='#ff6600')
plt.xticks(range(len(val_dict)))
plt.title('val')
plt.xlabel('Classes')
plt.ylabel('Number of videos')
plt.show()
```
